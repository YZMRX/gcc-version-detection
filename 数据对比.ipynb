#### 归一化处理

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
import numpy as np
std_scalar = StandardScaler()
std_scalar.fit(data1[s])
data1_std = std_scalar.transform(data1[s])
data1_std_df = pd.DataFrame(data1_std, columns=data1[s].columns, index=data1[s].index)
print('矩阵初值为：\n {}'.format(data1[s]))
print('该矩阵的均值为：{}\n 该矩阵的标准差为：{}'.format(std_scalar.mean_,np.sqrt(std_scalar.var_)))
data1_std_df.to_csv('data1_std.csv')
```

```python
from sklearn.preprocessing import StandardScaler
std_scalar = StandardScaler()
std_scalar.fit(data2[s])
data2_std = std_scalar.transform(data2[s])
data2_std_df = pd.DataFrame(data2_std, columns=data2[s].columns, index=data2[s].index)
print('矩阵初值为：\n {}'.format(data2[s]))
print('该矩阵的均值为：{}\n 该矩阵的标准差为：{}'.format(std_scalar.mean_,np.sqrt(std_scalar.var_)))
data2_std_df
```

#### pca主成分分析

```python
import sklearn.decomposition as dp
```

```python
pca = dp.PCA(n_components=3)
x1 = pca.fit_transform(data1_std_df)
print(pca.explained_variance_ratio_)
```

```python
x1 = pd.DataFrame(x1,columns=['feature1','feature2','feature3'])
df1 = pd.concat([x1,data1[s]['version']],axis=1)
df1
```

```python
x2 = pca.fit_transform(data2_std_df)
print('信息量大小：',pca.explained_variance_)
print('所占原始数据百分比：',pca.explained_variance_ratio_)
print('总占比：',pca.explained_variance_ratio_.sum())
data_pca1 = pd.DataFrame(pca.explained_variance_)
data_pca2 = pd.DataFrame(pca.explained_variance_ratio_)
data_pca = pd.concat([data_pca1,data_pca2],axis=1)
data_pca.columns = ['variance','variance_ratio']
data_pca.T.to_csv('data_pca.csv')
```

```python
x2 = pd.DataFrame(x2,columns=['feature1','feature2','feature3'])
df2 = pd.concat([x2,pd.DataFrame(data1[s]['version'])],axis=1)
df2
```

#### 模型预测

```python
from sklearn import tree
from sklearn.model_selection import train_test_split
```

```python
conlum_list = df1.columns.tolist() #分离特征和目标值
conlum_list.remove('version')
features = df1[conlum_list] #特征
target = df1['version'] # 什么版本
```

```python
model = tree.DecisionTreeClassifier(
    criterion="log_loss"
    ,max_depth=3
    ,splitter="random"
)
model.fit(features, target)
```

```python
predict1 = model.predict(df1.drop('version',axis=1))
# predict1 = pd.DataFrame(predict1,columns=['predict'])
# predict1 = pd.concat([df1,predict1],axis=1)
# predict1.to_csv('predict1.csv')
(predict1==df1['version']).mean()
```

```python
import graphviz
dot_data = tree.export_graphviz(model, out_file=None, feature_names=conlum_list, class_names=['6','7','8','9','10','11','12','13'],rounded=True, special_characters=True)
graph = graphviz.Source(dot_data)
graph.render('gcc')
```

```python
predict2 = model.predict(df2.drop('version',axis=1))
(predict2==df2['version']).mean()
```
