```python
# 返回指令集出现次数
import re
import collections
import pandas as pd
import os
def read_s(file_name):
    """
    读取指令集文件，并返回指令集出现次数的字典
    :param file_name: 指令集文件名
    :return: 指令集出现次数的字典
    """
    instruction_pattern = re.compile(r'\b(\w+)\b')
    dict_s = {}
    with open(file_name, 'r', encoding='utf-8') as f:
        lines = f.readlines()
        for line in lines:
            match1 = instruction_pattern.search(line)
            if  match1:
                instruction = match1.group(1)
                if dict_s.get(instruction) is None:
                    dict_s[instruction] = 1
                else:
                    dict_s[instruction] += 1
    return dict_s
```

```python
# 统计s文件各字符出现次数,保存为csv文件
def one_data(name:str,floder_name):
    # 输入的字典列表
    dict_list = []
    name_list = os.listdir(floder_name)
    for name_l in name_list:
        if name in name_l:
            temp = read_s(f"{floder_name}/{name_l}")
            temp["version"] = name_l.split("_")[1][:2]
            dict_list.append(temp)
    df = pd.DataFrame.from_dict(dict_list)
    df = df.fillna(0)
    df.to_csv(f'{name}.csv', index=False)  # 如果不希望保存索引列，可以将 index 参数设置为 False
```

构建SVM数学模型

给定一个样本数据集：
\begin{align}
S & = \{(x_{1},y_{2}),\cdots,(x_{i},y_{i})\}\in(X\times Y)^{n}
\end{align}
其中，$X$是样本的特征空间，$Y$是样本的标记空间，$n$是样本的个数，$R^{d}$表示$d$维实向量空间。

在该问题当中，S是一个线性不可分问题，我们使用以下约束条件进行求解，公式如下：
\begin{align}
\operatorname*{min}\frac{1}{2}||w||^{2}+C\sum_{i=1}^{n}\xi_{i}\quad s.t.\quad y_{i}(W^{T}x_{i}+b)\geq1-\xi_{i},\ \ (\xi_{i}\geq0,\ i=1,\cdots n)
\end{align}
其中，$W$是$R^{d}$上的$d$维向量，$b$是实数，$\xi_{i}$是松弛变量，$C$是正则化参数。

接着使用拉格朗日对偶性，求解上述的等价对偶问题，给定拉格朗日乘子向量
\begin{align}
\alpha=\left(\alpha_{1},\cdots,\alpha_{n}\right)^{T}
\end{align}
其中，$\alpha_{i}\geq0$，$\alpha_{i}$是拉格朗日乘子。

则对偶表达式为：

\begin{align}
\begin{cases}
\operatorname*{max}_{\alpha}\sum_{i  = 1}^{n}\alpha_{i}-\frac{1}{2}\sum_{i,j  = 1}^{n}\alpha_{i}\alpha_{j}y_{i}y_{j}k(x_{i},y_{j})\;\;s.t.\;\;\sum_{i  = 1}^{n}\alpha_{i}y_{i}= 0,\;\alpha_{i}\ge0,\;i = 1,\cdots,n
 \\
\operatorname*{max}_{\alpha}\sum_{i  = 1}^{n}\alpha_{i}-\frac{1}{2}\sum_{i,j  = 1}^{n}\alpha_{i}\alpha_{j}y_{i}y_{j}k(x_{i},y_{j})\;\;\;s t.\;\;\sum_{i  = 1}^{n}\alpha_{i}y_{i}= 0,\;0\leq\alpha_{i}\leq C,\;i = 1,\cdots,n
 \\
w  = \sum_{i  = 1}^{n}\alpha_{i}y_{i}x_{i}
 \\
b  = y_{j}-\sum_{i  = 1}^{n}\alpha_{i}y_{i}x_{i}^{T}x_{j}
\end{cases}
\end{align}

其中核函数$K(x,x_{i})$为，定义给定两个样本$x,x_{i}$有：
\begin{equation}K(x,x_i){=}\exp(-\gamma^*\|x-x_i\|^2)\end{equation}
其中，$\gamma$是核函数的参数，$c o e f$是核函数的偏置，$d$是核函数的维度。

经过上述处理之后，所以问题可转换为求$\alpha$的极大值即为求对偶问题的最优解，最终得出超平面和分类决策函数，如下所示：
\begin{align}
f(x)=\sum_{i=1}^{n}\alpha_{i}y_{i}k(x_{i},x)+b\;\;,\;\;b=y_{i}-\sum_{i=1}^{n}\alpha_{i}y_{i}k(x_{i},x_{j})
\end{align}

其中，$f(x)$的所在区域就是分类结果，也就是gcc的版本号
